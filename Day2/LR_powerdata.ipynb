```python
import numpy as np
filename = 'power.csv'
raw_data = open(filename,'r')
data = np.loadtxt(raw_data, delimiter =',')
data.shape
```

```python
data[:10]
```

```python
#Extracting Every column except last
X = data[:,:-1]
X.shape
```

```python
#extracting the last column
Y = data[:,-1]
Y.shape
```

```python
import matplotlib.pyplot as plt
#plotting the first feature against the dependent variable (Y) with the first 30 raws
X_30_1=X[:30,0]

X_30_1.shape 

Y_30=Y[:30]

plt.scatter(X_30_1, Y_30, color = 'red', s = 30)
```

```python
#plotting the second feature against the dependent variable (Y)
X_30_2=X[:30,1]

X_30_2.shape 

Y_30=Y[:30]

plt.scatter(X_30_2, Y_30, color = 'red', s = 30)
```

```python
#plotting the third feature against the dependent variable (Y)
X_30_3=X[:30,2]

X_30_3.shape 

Y_30=Y[:30]

plt.scatter(X_30_3, Y_30, color = 'red', s = 30)
```

```python
#plotting the third feature against the dependent variable (Y)
X_30_4=X[:30,3]

X_30_4.shape 

Y_30=Y[:30]

plt.scatter(X_30_4, Y_30, color = 'red', s = 30)
```

```python
X_train = X[:7500]
Y_train = Y[:7500]
X_test = X[7500:]
Y_test = Y[7500:]
```

```python
print(X_train.shape)
print(X_test.shape)
print(X_test.shape)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
Y_pred = regr.predict(X_test)
```

```python
print('Coefficients:\n', regr.coef_)
print(mean_squared_error(Y_test, Y_pred))
```

```python
plt.scatter(X_test[:,0], Y_test, color='black')
plt.plot(X_test[:,0], Y_pred, color = 'blue', linewidth = 3)
plt.xticks(())
plt.yticks(())
plt.show
```
